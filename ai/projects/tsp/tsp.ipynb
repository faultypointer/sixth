```python
def fitness(indi):
    cost = 0
    for curr, nex in zip(indi[:-1], indi[1:]):
        cost += data[curr][nex]
    return cost + data[indi[-1]][indi[0]]
```

```python
def crossover(p1: np.ndarray, p2: np.ndarray):
    o1 = p1.copy()
    o2 = p2.copy()
    S_no = np.random.randint(p1.size//2 - 1, p1.size//2 + 2)
    S_ind = np.random.choice(p1.size, S_no, replace=False)
    S_prime_ind = np.setdiff1d(np.arange(p1.size), S_ind)
    o1[S_ind] = p1[S_ind]
    o1[S_prime_ind] = p2[~np.isin(p2, p1[S_ind])]
    o2[S_ind] = p2[S_ind]
    o2[S_prime_ind] = p1[~np.isin(p1, p2[S_ind])]
    return o1, o2
```

```python
def mutation(indi):
    swap = np.random.choice(indi.size, 2)
    indi[swap] = indi[swap[::-1]]
```

```python
def tournament_selection(populations, fitness_scores, tournament_size=3):
    tournament_idx = np.random.choice(len(populations), tournament_size)
    tournament_fitness = [fitness_scores[i] for i in tournament_idx]
    winner_idx = tournament_idx[np.argmin(tournament_fitness)]  # min for TSP
    return populations[winner_idx]
```

```python
def elitism(populations, fitness_scores, num_elite=2):
    elite_idx = np.argsort(fitness_scores)[:num_elite]
    return populations[elite_idx]
```

```python
def adaptive_mutation_rate(gen, max_gen, initial_rate=0.001, final_rate=0.0001):
    return initial_rate - (initial_rate - final_rate) * (gen / max_gen)
```

```python
def genetic_algorithm(max_generations=1000, population_size=100, elite_size=2):
    best_solution = None
    best_fitness = float('inf')
    populations = np.array([np.random.permutation(len(data)) for _ in range(population_size)])    
    
    for gen in range(max_generations):
        # Calculate fitness for entire population
        population_fitness = np.array([fitness(x) for x in populations])
        current_mean_fitness = np.mean(population_fitness)
        
        # Store best solution
        gen_best_idx = np.argmin(population_fitness)
        gen_best_fitness = population_fitness[gen_best_idx]
        if gen_best_fitness < best_fitness:
            best_fitness = gen_best_fitness
            best_solution = populations[gen_best_idx].copy()
        
        # Early stopping
        if best_fitness < 1000:
            break
            
        # Elitism - preserve best solutions
        elite_solutions = elitism(populations, population_fitness, elite_size)
        
        # Create new population
        new_populations = np.zeros_like(populations)
        new_populations[:elite_size] = elite_solutions
        
        # Fill rest of population with crossover and mutation
        for i in range(elite_size, population_size, 2):
            # Tournament selection
            parent1 = tournament_selection(populations, population_fitness)
            parent2 = tournament_selection(populations, population_fitness)
            
            # Crossover
            offspring1, offspring2 = crossover(parent1, parent2)
            
            # Adaptive mutation
            mutation_rate = adaptive_mutation_rate(gen, max_generations)
            if np.random.random() < mutation_rate:
                mutation(offspring1)
            if np.random.random() < mutation_rate:
                mutation(offspring2)
                
            new_populations[i] = offspring1
            if i + 1 < population_size:
                new_populations[i + 1] = offspring2
        
        populations = new_populations
        
        # Logging
        if gen % 10 == 0:  # Print every 10 generations
            print(f"Generation: {gen}")
            print(f"Best Fitness: {best_fitness}")
            print(f"Mean Fitness: {current_mean_fitness}")
            print(f"Best Solution: {best_solution}")
            print("---")
    
    return best_solution, best_fitness
```
